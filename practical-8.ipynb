# Practical No.8: Write a Pandas program to create a line plot of the opening, closing stock prices of given company between two specific dates.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your stock price data into a Pandas DataFrame
# Replace 'your_data.csv' with the actual file containing your stock price data
df = pd.read_csv('/kaggle/input/nifty50-stock-market-data/HEROMOTOCO.csv')

# Convert the date column to a datetime data type (if not already)
df['Date'] = pd.to_datetime(df['Date'])

# Specify the date range
start_date = '2000-01-03'
end_date = '2000-03-13'

# Filter the data to include only the rows within the specified date range
filtered_df = df[(df['Date'] >= start_date) & (df['Date'] <= end_date)]

# Create a line plot of opening and closing prices
plt.figure(figsize=(12, 6))
plt.plot(filtered_df['Date'], filtered_df['Open'], label='Opening Price', color='blue')
plt.plot(filtered_df['Date'], filtered_df['Close'], label='Closing Price', color='red')

# Customize the plot
plt.title("HERO MOTOCORP STOCK DATA",fontsize=18)
plt.title('Stock Prices between {} and {}'.format(start_date, end_date))
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
